# Consolidated code to get required result

```python
import pandas as pd

#read data from souce to dataframe
df=pd.read_excel("H:\Powebi_learning\VehicleSales.xlsx",header=0)

df.columns=["Year","Bangalore_Car","Bangalore_Bike","Bangalore_Bi-Cycle",
            "Mumbai_Car","Mumbai_Bike","Mumbai_Bi-Cycle",
            "Chennai_Car","Chennai_Bike","Chennai_Bi-Cycle"]

#unpivot the column
df=df.melt(id_vars="Year",var_name="City_Vehicle",value_name="Sales")

# split the column to create 2 columns in DataFrame
df[["City","Vehicle"]]=df["City_Vehicle"].str.split("_",expand=True)

df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')

#Pivot the dataframe to get in well structured format
df_pivot=df.pivot_table(index=["Year","City"],columns="Vehicle",values="Sales").reset_index()

# redefine the column heading
df_pivot.columns.name=None
df_pivot.columns=["Year","City","Bi-Cycle","Bike","Car"]

#formatted clean Dataframe
df_pivot
```
