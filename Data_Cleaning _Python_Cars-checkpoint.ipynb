```python
import pandas as pd

# Convert the dictionary to DataFrame
df=pd.read_excel("VehicleSales.xlsx")

# Set the first row as column headers
df.columns = df.iloc[0]
df = df[1:]

# Renaming columns for simplicity
df.columns = ['Year', 'Bangalore_Car', 'Bangalore_Bike', 'Bangalore_Bi-Cycle',
              'Mumbai_Car', 'Mumbai_Bike', 'Mumbai_Bi-Cycle',
              'Chennai_Car', 'Chennai_Bike', 'Chennai_Bi-Cycle']

# Melting the DataFrame to long format
df_melted = pd.melt(df, id_vars=['Year'], var_name='City_Vehicle', value_name='Sales')

# Splitting the 'City_Vehicle' column into 'City' and 'Vehicle' columns
df_melted[['City', 'Vehicle']] = df_melted['City_Vehicle'].str.split('_', expand=True)

# Pivoting the table to the desired format
df_final = df_melted.pivot_table(index=['City', 'Year'], columns='Vehicle', values='Sales').reset_index()

# Renaming the columns to the desired format
df_final.columns.name = None
df_final.rename(columns={'Car': 'Sale_of_Car', 'Bike': 'Sale_of_Bike', 'Bi-Cycle': 'Sale_of_Bi-cycle'}, inplace=True)

df_final
```

```python
import seaborn as sn
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Melt the DataFrame for plotting
df_plot = df_final.melt(id_vars=['City', 'Year'], value_vars=['Sale_of_Bi-cycle', 'Sale_of_Bike', 'Sale_of_Car'], 
                        var_name='Vehicle_Type', value_name='Sales')

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(data=df_plot, y='City', x='Sales', hue='Vehicle_Type')
plt.title('Sales of Bicycles, Bikes, and Cars by City')
plt.show()
```
